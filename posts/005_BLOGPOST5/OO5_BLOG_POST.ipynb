Today I played around with ChatGPTs ability to do simple but intricite tasks. I wanted to test the LLMs ability to do a long drawn out task without getting off task or becoming dilusional. So I started by giving ChatGPT an email outline and list of 100 randomly generated names. I then asked GPT to address the email to the 100 names and return me the list so I could copy and paste each one into an email. GPT was remarkably good at this, it returned me the first 10 names and then asked if I wanted the entire list converted into a CSV file, and Excel file, or PDF to save me from scrolling through the the entire 100 names. I first tried the CSV file. I used Google Colab to open the CSV file and filter the CSV for 100 unique names and saw that GPT had done what I asked it perfectly and did not miss a name.

I then wanted to check GPTs ability to do the same task with less instruction so I asked GPT to make a unique email and address them to the same 100 people. The original email I had given GPT was a networking email, my hope was that GPT would pick up on the context, and goal of the email and recreate this in a unique way for each person. 

GPT created 100 emails that at a glance seemed individually unique, but after reading through them GPT seems to have created a bucket of emails and used the same 10 or so emails over again with slight alterations to the grammer, syntax and word choice.
