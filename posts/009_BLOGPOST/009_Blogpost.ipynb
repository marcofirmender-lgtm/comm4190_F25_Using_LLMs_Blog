I started by trying to use Claude for this expiriment but I had no tokens available so I moved over to ChatGPT. I started by telling GPT that I was going to need help creating STL files, which is the industry standard file format for 3D printing. once GPT was familiar with what I would be asking it to do I moved onto prompting the AI.

My first prompt was: 

*I'd like you to create me a 2-foot-tall cat with a light saber and a cape, make the cat a Tabby cat, and the light saber should be red.*

GPT took a couple minutes to design the cat and set up the STL model which was not fully shocking but a bit unexpected. This was partially due to the complexity of my prompt; creating an STL with multiple colors and shapes can be tricky. Once I opened the file on TinkerCAD, I found that the design specs I had outlined were not met.
